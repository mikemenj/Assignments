```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
os.makedirs('Data',exist_ok=True)
```

```python
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
akas_url="https://datasets.imdbws.com/title.akas.tsv.gz"
ratings_url="https://datasets.imdbws.com/title.ratings.tsv.gz"
```

# AKAs

```python
%%time
# title akas
chunk_num = 1
df_reader = pd.read_csv(akas_url, sep='\t',
                        low_memory=False, chunksize=100_000)
```

```python
%%time
for temp_df in df_reader:
      
    ## only us movies
    temp_df = temp_df[temp_df['region']=='US']
    
    ## Replace "\N" with np.nan
    temp_df.replace({'\\N':np.nan},inplace=True)
        
    ### Saving chunk to disk
    fname= f'Data/title_akas_chunk_{chunk_num:03d}.csv.gz'
    temp_df.to_csv(fname, compression='gzip')
    print(f"- Saved {fname}")
    
    ## increment chunk_num    
    chunk_num+=1

    ## Closing the reader now that we are done looping through the file
df_reader.close()
```

```python
# Recursive query - extra /**/ added to string
q = "Data/**/title_akas_chunk*.csv.gz" 
chunked_files = sorted(glob.glob(q, recursive=True)) 
# Showing the first 5 
chunked_files[:5]
```

```python
## Loading and Concatenating the list of dfs with 1 line
akas_combined = pd.concat([pd.read_csv(file, index_col=0) for file in chunked_files])
akas_combined.head()
```

```python
## Save current dataframe to file.
akas_combined.to_csv("Data/title_akas.csv.gz",compression='gzip',index=False)
```

```python
# Open saved file and preview again
akas_combined = pd.read_csv("Data/title_akas.csv.gz", low_memory = False)
akas_combined.head()
```

```python
akas_combined.info()
```

# Basics

```python
%%time
# title basics
basics = pd.read_csv(basics_url, sep='\t')
```

```python
basics.info()
basics.head()
```

```python
basics = basics[(basics['startYear']=='2000')|(basics['startYear']=='2021')|(basics['startYear']=='2022')]
basics.info()
```

```python
basics['startYear'].astype('int')
```

```python
basics.replace({'\\N':np.nan},inplace=True)
```

```python
basics.isnull().sum()
```

```python
basics = basics.dropna(subset=['runtimeMinutes','genres']) 
basics.info()
```

```python
basics = basics[basics['titleType']=='movie']
basics.info()
```

```python
# Exclude movies that are included in the documentary category.
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
basics.info()
```

```python
keepers = basics['tconst'].isin(akas_combined['titleId'])
basics = basics[keepers]
basics.info()
```

```python
## Save current dataframe to file.
basics.to_csv("Data/title_basics.csv.gz",compression='gzip',index=False)
```

```python
# Open saved file and preview again
basics = pd.read_csv("Data/title_basics.csv.gz", low_memory = False)
basics.info()
basics.head(3)
```

# Ratings

```python
%%time
# title ratings
ratings = pd.read_csv(ratings_url, sep='\t')
ratings.info()
ratings.head(3)
```

```python
## Replace "\N" with np.nan
ratings.replace({'\\N':np.nan},inplace=True)
```

```python
keepers = ratings['tconst'].isin(akas_combined['titleId'])
ratings = ratings[keepers]
ratings.info()
ratings.head(3)
```

```python
## Save current dataframe to file.
ratings.to_csv("Data/title_ratings.csv.gz",compression='gzip',index=False)
```

```python
# Open saved file and preview again
ratings = pd.read_csv("Data/title_ratings.csv.gz", low_memory = False)
ratings.info()
ratings.head(3)
```
