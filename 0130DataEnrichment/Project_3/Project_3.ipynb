```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
os.makedirs('Data',exist_ok=True)
```

```python
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
akas_url="https://datasets.imdbws.com/title.akas.tsv.gz"
ratings_url="https://datasets.imdbws.com/title.ratings.tsv.gz"
```

# akas

```python
%%time
# title akas
chunk_num = 1
df_reader = pd.read_csv(akas_url, sep='\t',
                        low_memory=False, chunksize=100_000)
```

```python
%%time
for temp_df in df_reader:
        #### COMBINED WORKFLOW FROM ABOVE
    ## only us movies
    temp_df = temp_df[temp_df['region']=='US']
    
    ## Replace "\N" with np.nan
    temp_df.replace({'\\N':np.nan},inplace=True)
        
    ### Saving chunk to disk
    fname= f'Data/title_akas_chunk_{chunk_num:03d}.csv.gz'
    temp_df.to_csv(fname, compression='gzip')
    print(f"- Saved {fname}")
    
    ## increment chunk_num    
    chunk_num+=1

    ## Closing the reader now that we are done looping through the file
df_reader.close()

```

```python
# Recursive query - extra /**/ added to string
q = "Data/**/title_akas_chunk*.csv.gz" 
chunked_files = sorted(glob.glob(q, recursive=True)) 
# Showing the first 5 
chunked_files[:5]
```

```python
## Loading and Concatenating the list of dfs with 1 line
akas_combined = pd.concat([pd.read_csv(file, index_col=0) for file in chunked_files])
akas_combined
```

```python
## Save current dataframe to file.
akas_combined.to_csv("Data/title_akas.csv.gz",compression='gzip',index=False)
```

```python
# Open saved file and preview again
akas_combined = pd.read_csv("Data/title_akas.csv.gz", low_memory = False)
akas_combined.head()
```

# Basics

```python
%%time
# title basics
chunk_num = 1
df_reader = pd.read_csv(basics_url, sep='\t',
                        low_memory=False, chunksize=100_000)

```

```python
%%time
for temp_df in df_reader:
        #### COMBINED WORKFLOW FROM ABOVE
    ## Replace "\N" with np.nan
    temp_df.replace({'\\N':np.nan},inplace=True)
    
    ## Eliminate movies that are null for runtimeMinute, genres, and startYear
    temp_df = temp_df.dropna(subset=['runtimeMinutes','genres','startYear']) 
        
    ## NOTE: THERE ARE ADDITIONAL REQUIRED FILTERING STEPS FOR THE PROJECT NOT SHOWN HERE
    
    ### Convert startyear to numeric for slicing
    
    ## convert numeric features
    temp_df['startYear'] = temp_df['startYear'].astype(float)
    
    ## keep startYear 2000-2022
    temp_df = temp_df[(temp_df['startYear']>=2000)&(temp_df['startYear']<2022)]
    
    # Exclude movies that are included in the documentary category.
    is_documentary = temp_df['genres'].str.contains('documentary',case=False)
    temp_df = temp_df[~is_documentary]

    is_movie = temp_df['titleType']=='movie'
    temp_df = temp_df[is_movie]
    
    ### Saving chunk to disk
    fname= f'Data/title_basics_chunk_{chunk_num:03d}.csv.gz'
    temp_df.to_csv(fname, compression='gzip')
    print(f"- Saved {fname}")
    
    ## increment chunk_num    
    chunk_num+=1

    ## Closing the reader now that we are done looping through the file
df_reader.close()

```

```python
# Recursive query - extra /**/ added to string
q = "Data/**/title_basics_chunk*.csv.gz" 
chunked_files = sorted(glob.glob(q, recursive=True)) 
# Showing the first 5 
chunked_files[:5]
```

```python
## Loading and Concatenating the list of dfs with 1 line
basics_combined = pd.concat([pd.read_csv(file, index_col=0) for file in chunked_files])
basics_combined
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers =basics_combined['tconst'].isin(akas_combined['titleId'])
keepers
```

```python
basics_combined = basics_combined[keepers]
basics_combined
```

```python
## Save current dataframe to file.
basics_combined.to_csv("Data/title_basics.csv.gz",compression='gzip',index=False)
```

```python
# Open saved file and preview again
basics_combined = pd.read_csv("Data/title_basics.csv.gz", low_memory = False)
basics_combined.head()
```
