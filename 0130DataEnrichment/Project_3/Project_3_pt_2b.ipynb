```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
path = 'Data/final_tmdb_data_2000.csv.gz'
df1 = pd.read_csv(path)
df1.info()
df1.head()
```

```python
path = 'Data/final_tmdb_data_2001.csv.gz'
df2 = pd.read_csv(path)
df2.info()
df2.head()
```

```python
df = pd.concat([df1,df2]).reset_index(drop=True)
df.to_csv(f"Data/tmdb_results_combined.csv.gz", compression="gzip", index=False)
df.info()
```

```python
path = 'Data/tmdb_results_combined.csv.gz'
df = pd.read_csv(path)
df.info()
df.head()
```

# EDA

## #1

```python
budget_filter = df['budget']>0
revenue_filter = df['revenue']>0
print(f'{len(df[budget_filter|revenue_filter])} movies had some valid financial information (values > 0 for budget OR revenue)')
```

```python
df = df[budget_filter|revenue_filter].reset_index(drop=True)
df.info()
```

## #2

```python
certs = df.groupby('certification').size().sort_values(ascending = False)
certs
```

```python
certs.plot(kind='bar')

plt.title('Number of Movies Grouped by Rating')
plt.xlabel('Movies')
plt.ylabel('Count')

plt.show()
```

## #3

```python
avg_rev = df.groupby('certification')['revenue'].mean().sort_values(ascending = False)
avg_rev.map("${:,.2f}".format)
```

```python
avg_rev.plot(kind='bar')

plt.title('Average Revenue of Movies Grouped by Rating')
plt.xlabel('Movies')
plt.ylabel('Average Revenue')

plt.show()
```

## #4

```python
avg_bud = df.groupby('certification')['budget'].mean().sort_values(ascending = False)
avg_bud.map("${:,.2f}".format)
```

```python
avg_bud.plot(kind='bar')

plt.title('Average budget of Movies Grouped by Rating')
plt.xlabel('Movies')
plt.ylabel('Average budget')

plt.show()
```
