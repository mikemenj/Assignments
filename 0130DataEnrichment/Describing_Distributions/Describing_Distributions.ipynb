```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from scipy import stats
```

```python
plt.style.use(['seaborn-muted','fivethirtyeight'])
sns.set_context('talk')
path ="https://docs.google.com/spreadsheets/d/1APV3pXiAszS_0mSgkiEt9IUNH-QmyX7KwxSAwuADl6Y/gviz/tq?tqx=out:csv&sheet=medical_data"
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

# VitD_Levels

```python
col = 'VitD_levels'
```

```python
sns.kdeplot(data = df, x= col);
```

```python
# Find the mean, median
mean = df[col].mean()
median = df[col].median()
std = df[col].std()
plus_one_std = mean + std
minus_one_std = mean - std
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(df[col], ax = ax, stat='probability')
ax.axvline(mean,color = 'red',label=f'Mean = {mean:,.2f}')
ax.axvline(median, ls='--', color = 'green', label = f'Median = {median:,.2f}')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(col)
ax.legend();
```

```python
kurt = stats.kurtosis(df[col], fisher = False)
kurt
```

- Is it Discrete or Continuous? Continuous
- Does it have a skew? no
- What type of kurtosis does it display? Mesokurtic
