```python
# Exploring existing format with a few examples
df1['Hero|Publisher'].head(2)
```

```python
## save the 2 new columns into the dataframe
df1[['Hero','Publisher']] = df1['Hero|Publisher'].str.split('|',expand=True)
df1.head(2)
```

```python
## drop the original column 
df1 = df1.drop(columns=['Hero|Publisher'])
df1.head(2)
```

```python
## examining a single value from the coordinates col
measurements = df1.loc[0,"Measurements"]
print(type(measurements))
measurements
```

```python
measurements = measurements.replace("'",'"') 
measurements
```

```python
json.loads(measurements)
```

```python
## use .str.replace to replace all single quotes
df1['Measurements'] = df1['Measurements'].str.replace("'",'"')

## Apply the json.loads to the full column
df1['Measurements'] = df1['Measurements'].apply(json.loads)
df1['Measurements'].head()
```

```python
#use pandas pd.Series function on our column of dictionaries will automatically convert each key to a new column
hw = df1['Measurements'].apply(pd.Series)
hw
```

```python
# concat long_lat with original dataframe
df1 = pd.concat((df1, hw), axis = 1)
df1.head(2)
```

```python
df1 = df1.drop(columns=['Measurements'])
df1.head()
```

```python
## use .str.replace to replace all single quotes and cast as numeric/float
df1['Height'] = df1['Height'].str.replace(" cm",'').astype('float')
df1['Height'].head()
```

```python
## use .str.replace to replace all single quotes and cast as numeric/float
df1['Weight'] = df1['Weight'].str.replace(" kg",'').astype('float')
df1['Weight'].head()
```
